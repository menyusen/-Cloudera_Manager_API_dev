```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
```

```python
f_cpu = open('./Pro_HSY_data/cpu_percent_used.out','r+')
f_memory = open('./Pro_HSY_data/total_physical_memory_used.out','r+')
f_capacity = open('./Pro_HSY_data/capacity_used.out','r+')
f_yarn_cpu = open('./Pro_HSY_data/yarn_cpu_used.out','r+')
f_yarn_memory = open('./Pro_HSY_data/yarn_memory_used.out','r+')
```

```python
# Cluster CPU Usage (%)
k1=0
plt.figure(1)
y_cpu = []
time1 = []
for line in f_cpu:
   dp = line.split('\t')
   time1.append(dp[0])
   y_cpu.append(dp[1])
   k1 += 1

x1 = []
print "num of CPU sampling point:",k1
for i in range(k1):
   x1.append(i)
#print x,y  
#x = np.arange(0, 72, 1)
ylim(1, 100)
plt.title('Cluster CPU Usage') 
plt.xlabel(time1[0]+' - '+time1[k1-1])
plt.ylabel('cpu_percent_used (%)')
plt.plot(x1,y_cpu)
f_cpu.close()
```

```python
#Cluster Memory Usage
k2=0
plt.figure(2)
time2=[]
y_memory = []
y_memory_p = []
for line in f_memory:
   dp = line.split('\t')
   #x.append(dp[0])
   time2.append(dp[0])
   y_memory.append(dp[1])
   y_memory_p.append(float(dp[1])/26.8*100)
   k2 += 1

x2 = []
print "num of memory sampling point:",k2
for i in range(k2):
   x2.append(i)
#print x,y  
#x = np.arange(0, 72, 1)
#ylim(1, 10)

#Cluster Memory Usage (TB)
p21 = plt.subplot(211)
p21.set_title('Cluster Memory Usage')
#p21.set_xlabel(time2[0]+' - '+time2[k2-1])
p21.set_ylabel('total_physical_memory_used (TB)')
p21.plot(x2,y_memory)

#Cluster Memory Percent Usage (%)
p22 = plt.subplot(212)
#p22.set_title('Cluster Memory Usage')
p22.set_xlabel(time2[0]+' - '+time2[k2-1])
p22.set_ylabel('total_physical_memory_percent_used (%)')
p22.plot(x2,y_memory_p)
f_memory.close()
```

```python
#Cluster Capacity Usage
k3 = 0    
plt.figure(3)
time3 = []
y_capacity = []
y_capacity_p = []
for line in f_capacity:
   dp = line.split('\t')
   #x.append(dp[0])
   time3.append(dp[0])
   y_capacity.append(dp[1])
   y_capacity_p.append(float(dp[1])/5.2*100)
   k3 += 1

x3 = []
print "num of capacity sampling point:",k3
for i in range(k3):
   x3.append(i)
#print x,y  
#x = np.arange(0, 72, 1)
#ylim(3, 5.2)

#Cluster Capacity Usage (PB)
p31 = plt.subplot(211)
p31.set_title('Cluster HDFS Capacity Usage')
#p31.set_xlabel(time3[0]+' - '+time3[k3-1])
p31.set_ylabel('capacity_used (PB)')
p31.plot(x3,y_capacity)

#Cluster Capacity Percent Usage (%)
p32 = plt.subplot(212)
#p32.set_title('Cluster Capacity Usage')
p32.set_xlabel(time3[0]+' - '+time3[k3-1])
p32.set_ylabel('capacity_percent_used (%)')
p32.plot(x3,y_capacity_p)
f_capacity.close()
```

```python
#YARN CPU Usage
k4 = 0
plt.figure(4)
time4 = []
y_yarn_cpu = []
y_yarn_cpu_p = []
for line in f_yarn_cpu:
   dp = line.split('\t')
   #x.append(dp[0])
   time4.append(dp[0])
   y_yarn_cpu.append(dp[1])
   y_yarn_cpu_p.append(float(dp[1])/1800*100)
   k4 += 1

x4 = []
print "num of YARN CPU sampling point:",k4
for i in range(k4):
   x4.append(i)
#print x,y  
#x = np.arange(0, 72, 1)
#ylim(3, 5.2)

##YARN CPU Usage(C)
p41 = plt.subplot(211)
p41.set_title('YARN CPU Usage')
#p41.set_xlabel(time4[0]+' - '+time4[k4-1])
p41.set_ylabel('yarn_cpu_used (c)')
plt.plot(x4,y_yarn_cpu)

##YARN CPU Usage(%)
p42 = plt.subplot(212)
#p42.set_title('YARN CPU Usage')
p42.set_xlabel(time4[0]+' - '+time4[k4-1])
p42.set_ylabel('yarn_cpu_percent_used (%)')
plt.plot(x4,y_yarn_cpu_p)
f_yarn_cpu.close()
```

```python
#YARN Memory Usage
k5 = 0
plt.figure(5)
time5 = []
y_yarn_memory = []
y_yarn_memory_p = []
for line in f_yarn_memory:
   dp = line.split('\t')
   #x.append(dp[0])
   time5.append(dp[0])
   y_yarn_memory.append(dp[1])
   y_yarn_memory_p.append(float(dp[1])/8.62*100)
   k5 += 1

x5 = []
print "num of YARN memory sampling point:",k5
for i in range(k5):
   x5.append(i)
#print x,y  
#x = np.arange(0, 72, 1)
#ylim(3, 5.2)

#YARN Memory Usage(TB)
p51 = plt.subplot(211)
p51.set_title('YARN Memory Usage')
#p51.set_xlabel(time5[0]+' - '+time5[k5-1])
p51.set_ylabel('yarn_memory_used (TB)')
p51.plot(x5,y_yarn_memory)

#YARN Memory Percent Usage (%)
p52 = plt.subplot(212)
p52.set_xlabel(time5[0]+' - '+time5[k5-1])
p52.set_ylabel('yarn_memory_percent_used (%)')
p52.plot(x5,y_yarn_memory_p)
f_yarn_memory.close()
```
